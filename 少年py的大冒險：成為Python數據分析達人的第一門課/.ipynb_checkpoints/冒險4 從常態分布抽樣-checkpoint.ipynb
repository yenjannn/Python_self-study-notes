````python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
````

利用`numpy(np)`下的子套件叫做`random`，在`random`下有個叫
```Python
randn
```
的指令，這是用來從標準常態分布隨機抽樣取亂數。

````python
np.random.randn(10)
````

### 畫一百個點

````python
plt.plot(np.random.randn(100))
````

````python
X = [0, 1.1, 2.5, 3.2, 4.8]
Y = [2.1, 3.4, 1, 5, 7]
````

````python
plt.scatter(X, Y)
````

X、Y的座標當然也可以都從標準常態分布出來。

````python
X = np.random.randn(300)
Y = np.random.randn(300)

plt.scatter(X, Y)
````

給定一個區間，隨機找出一個整數。      
```Python
np.random.randint(n, m)
```
亂數找出某個正數r，符合$n\leq r<m$。

````python
r = np.random.randint(1, 101)
````

````python
r
````

### 冒險旅程4

從平均值2，標準差5符合常態分布的資料中，取出100筆資料。
[參考資料](https://docs.scipy.org/doc/numpy-1.15.1/reference/generated/numpy.random.randn.html)

````python
plt.plot(np.sqrt(5)*np.random.randn(100) + 2)
````
