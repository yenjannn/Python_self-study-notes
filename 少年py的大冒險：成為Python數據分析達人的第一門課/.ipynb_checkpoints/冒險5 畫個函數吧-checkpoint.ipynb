電腦畫函數的方式，基本上就是我們小時候學過的「描點法」。
現在我們來畫$f(x)=x^2+1$。
1. 決定x座標範圍，比如說-10到10
2. 然後取500個點。
```Python
np.linspace(開始, 結束, n)
```

````python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
````

````python
X = np.linspace(-10, 10, 500)
Y = X**2 + 1

plt.plot(X, Y)
````

$f(x)=sin(x)$，範圍取[0,6$\pi$]，1000個點。

````python
π = np.pi
````

````python
θ = np.linspace(0, 6*π, 1000)
````

````python
y = np.sin(θ)
````

````python
plt.plot(θ, y)
````

### 冒險旅程5

找一個覺得很漂亮的函數，畫出它的圖形。

我想畫$y=\frac{sin(x)}{x}$，範圍取[0,6$\pi$]，1000個點。

````python
π = np.pi
X = np.linspace(-6*π, 6*π, 1000)
y = np.sin(X)/X
````

````python
plt.plot(X, y)
````
