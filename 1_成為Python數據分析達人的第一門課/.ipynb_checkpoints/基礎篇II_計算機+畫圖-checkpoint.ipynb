#### 以後每次用Python時一開始都要打入的一串字。
```Python
%matplotlib inline
```
這裡的百分比符號其實不是Python的指令，而是Jupyter Notebook特有的「魔術指令」。意思是要求畫出來的圖直接顯示在網頁中，而不要另外跳出一個視窗。
```Python
import numpy as np
```

#### 引入套件的標準方法，並自訂(約定俗成)一個套件的簡稱。同理matplotlib也可以這樣引進。
```Python
import matplotlib.pyplot as plt
```
注意matplotlib是很大的套件庫，而一般我們只是要基本畫圖用的那個叫pyplot的子套件庫，所以我們只讀進來它！並且自訂(又是約定俗成)命名為plt。

#### 使用matplotlib.pyplot最主要的兩種畫圖方式。

折線圖
```Python
plt.plot
```
散佈圖
```Python
plt.scatter
```

````python
# 試試看
plt.plot([2, -5, 3, 8, -2])
````

上圖畫的是：     
(0,2),(1,-5),(2,3),(3,8),(4,-2)      
這幾個點，然後連起來。也就是我們打入的一串數字 [2,-5,3,8,-2]      
其實是這些點的y座標，用中括弧括起來是Python的串列(list)資料型態。串列式Python裡非常重要的資料型態，基本上就是一堆數據放在一起。

#### 標準的畫圖方式
最標準的畫圖方式是把我們數據點的x座標用一個串列收集起來，比方說叫X；然後再把y座標收集起來，叫Y。然後用`plt.plot(X, Y)`把圖畫出來。
我們的例子中，只給了y座標，這時Python會自動補上x座標 [0,1,2,3,4]，因此我們需要把x座標、y座標分開，假設我們分別用X和Y兩個變數來接。

````python
X = [0, 1, 2, 3, 4]
````

````python
Y = [2, -5, 3, 8, -2]
````

````python
plt.plot(X, Y)
````

### 5. 畫個函數吧
電腦畫函數的方式，基本上就是我們小時候學過的「描點法」。
現在我們來畫$f(x)=x^2+1$。
1. 決定x座標範圍，比如說-10到10
2. 然後取500個點。
```Python
np.linspace(開始, 結束, n)
```

````python
X = np.linspace(-10, 10, 500)
Y = X**2 + 1

plt.plot(X, Y)
````

$f(x)=sin(x)$，範圍取[0,6$\pi$]，1000個點。

````python
π = np.pi
````

````python
θ = np.linspace(0, 6*π, 1000)
````

````python
y = np.sin(θ)
````

````python
plt.plot(θ, y)
````

### 6. 參數式畫圖其實也一樣！

````python
π = np.pi
θ = np.linspace(0, 2*π, 500)
````

````python
r = 3
x = r*np.cos(θ)
y = r*np.sin(θ)
````

````python
plt.plot(x, y)
````

````python
ax = plt.gca()
ax.set_aspect('equal')

plt.plot(x, y)
````

````python
r = 1 - np.sin(θ)

x = r*np.cos(θ)
y = r*np.sin(θ)
````

````python
ax = plt.gca()
ax.set_aspect('equal')

plt.plot(x, y)
````

````python
ax = plt.gca()
ax.set_aspect('equal')

plt.plot(x, y, c='r')
````
