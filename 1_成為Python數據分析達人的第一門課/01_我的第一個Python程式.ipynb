### 定義一個函數
可重複利用。   
shift+tab兩次:出現解說。

```python
def myplot(n=1):    #n=1 預設值為1
    x = np.linspace(0, 10, 200)
    y = np.sin(n*x)
    plt.plot(x,y)
```

```python
myplot(3)
```

```python
myplot()
```

```python
from ipywidgets import interact
```

```python
interact(myplot, n=(1, 100))
```
