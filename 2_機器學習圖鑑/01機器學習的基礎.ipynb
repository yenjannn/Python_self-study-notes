### 監督式學習(分類)範例

```python
from sklearn.datasets import load_breast_cancer
```

```python
data1 = load_breast_cancer()
```

```python
print(data1.DESCR)
```

```python
X = data1.data
y = data1.target
```

```python
data1.data
```

```python
len(y)
```

```python
X = X[:, :10]
```

```python
X
```

Losgistic Regression羅吉斯迴歸

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(X, y)
```

```python
y_pred = model.predict(X)
```

模型評估

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y, y_pred)
```

### 非監督式學習分群範例

```python
from sklearn.datasets import load_wine
```

```python
data2 = load_wine()
```

```python
data2
```

```python
X = pd.DataFrame(data2.data, columns=data2.feature_names)
y = pd.DataFrame(data2.target, columns=['Kind(target)'])

df2 = pd.concat([X, y], axis=1)
```

```python
df2.head()
```

```python
X = data2.data[:, [0, 9]]
```

```python
X
```

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters=3)
```

```python
pred = model.fit_predict(X)
```

```python
pred
```

```python
plt.scatter(X[:, 0], X[:, 1], c=pred, marker='s')
```

```python
#製作figure
fig = plt.figure()

#設定axes
ax = fig.add_subplot(1, 1, 1)

#方形
ax.scatter(X[:, 0], X[:, 1], c=pred, marker='s')

#圖例，標題等
ax.set_title('Sample Title')
ax.legend(['labe1', 'label2', 'label3'])  # label進不去

plt.show()
```
