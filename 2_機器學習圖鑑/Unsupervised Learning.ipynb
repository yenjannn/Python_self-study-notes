### 01 PCA(Principal Components Analysis)

```python
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
```

```python
data = load_iris()
```

```python
print(data.DESCR)
```

```python
model = PCA(n_components=2)
```

```python
model = model.fit(data.data)
```

```python
print(model.transform(data.data))
```
